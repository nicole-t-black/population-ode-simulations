To limit the population, we can introduce K, or carrying capacity, which represents the max population that can be supported given the constrainst of the environment. K is integrated into the Malthusian equation with the equation $\frac{dN}{dt} = \alpha N(1 - \frac{N}{K})$.

Here, $\frac{dN}{dt}$ represents the change in population over time, $\alpha N$ represents the growth rate given infinite resources, and $1- \frac{N}{K}$ represents positive or negative difference bewteen the current population and the max population supported, K. When N >> K, this term is a large negative, and so the change in population will decrease quickly. When K >> N, this term is positive and very close to one, so the growth will be modeled by $\alpha N$. If K and N are approximately equal, the term will be near 0, so growth/decay will be small or nonexistant.

5. Simply put, Hebbs rule states that neurons that fire together, wire together. More precisely, the synaptic efficacy between sets of neurons increases the more frequently that set of neurons fires simultaneously. Consider a simple example of Hebbs rule consisting of two neurons firing one postsynaptic neuron:

Here, $x_1, x_2$ give the activity/firing levels of the input (presynaptic) neurons, and $y$ gives the activity/firing level of the output (postsynaptic) neuron. $w_1, x_2$ are the efficacies between the pairs of neurons as shown above. We will assume that the values of $x_1, x_2$, are always positive (the neurons $x_1, x_2$ fire) and the values $w_1, w_2$ are always non-negative (meaning $x_1$ and $x_2$ cannot inhibit neuron $y$ from firing). According to Hebbs rule, the more frequently $x_1$ and $y$ fire together, the larger the value $w_1$ will become, and likewise for $w_2$. We represent this as mathematical model described by the ODEs:

 $\frac{dw_i}{dt} = x_iy$ for $i=1,2$ where the firing of neuron y is given by $y = x_1w_1+x_2w_2$.

a. Solve to find the fixed point of this system. 

// set the derivatives equal to 0

$\frac{dw_1}{dt} = x_1y = 0$

$\frac{dw_2}{dt} = x_2y = 0$


// solve for y

$\frac{dw_1}{dt} = x_1y = 0$ // divide by $x_1$

$y = \frac{0}{x_1} = 0$

$y = 0$

// plug in

$y = 0 = x_1w_1+x_2w_2$

$x_1w_1 = -x_2w_2$

Given that $x_1$ and $x_2$ must be positive, for $x_1w_1 = -x_2w_2$, $w_1$ and $w_2$ must be 0.

Our fixed point is $(w_1, w_2) = (0,0)$

b. Is the fixed point stable or unstable? Explain your reasoning.

Let $\frac{dw_1}{dt} = 0$, $\frac{dw_2}{dt} = x_1(x_1w_1+x_2w_2)$

Solve $\frac{dw_2}{dt}|_{w_2=0+\alpha}$, where $\alpha$ is a positive pertubation of $w_2$.

$\frac{dw_2}{dt}|_{w_2=0+\alpha} = x_1x_2\alpha$

Since $x_1,x_2,$ and $\alpha$ are always positive, $\frac{dw_2}{dt}|_{w_2=0+\alpha}$ is also positive. Hence, given a pertubation of $w_2$ in the positive direction, the derivative $\frac{dw_2}{dt}$ grows, rather than returning to 0, and the fixed point is unstable.

We do not need to do the above calculations for a pertubation in $w_1$ since the fixed point is already unstable in one direction (but if we did do it, we would find the same thing).

c. Consider a modified model where the value of $w_1$ is bounded by a constant parameter, $\gamma$. The new model is:

$\frac{dw_i}{dt} = x_iy(\gamma-w_i)$, for $i=1,2$

Solve to find the fixed point(s) of this system. 

// set the derivatives equal to 0

$\frac{dw_1}{dt} = x_1y(\gamma-w_1) = 0$

$\frac{dw_2}{dt} = x_2y(\gamma-w_2) = 0$

// solve for y

$x_1y(\gamma-w_1) = 0$ // divide by $x_1$

$y(\gamma-w_1) = 0$ // multiply out

$y\gamma - yw_1 = 0$ // add $yw_1$ to both sides

$y\gamma = yw_1$

For the above equation to be true, either y must be 0 or $w_1$ must be $\gamma$.

If y = 0, $x_1w_1+x_2w_2$ = 0, which only occurs at $(w_1,w_2) = (0,0)$

Looking at the other equation:

$\frac{dw_2}{dt} = x_2y(\gamma-w_2) = 0$ // divide by $x_2$

$y(\gamma-w_2) = 0$ // multiply out

$y\gamma-yw_2 = 0$ // add $yw_2$ to both sides

$y\gamma = yw_2$

Again, for the above equation to be true, either y must be 0 or $w_2$ must be $\gamma$

Hence, the fixd points fo the system are $(w_1,w_2) = (0,0)$ and $(w_1,w_2) = (\gamma,\gamma)$

d. Are the fixed points of the new model stable or unstable? Explain your reasoning. Describe in words what this means in terms how the neurons are firing.

Let $\frac{dw_1}{dt} = 0, \frac{dw2}{dt} = x_2y(\gamma - w_2) = x_2(x_1w_1+x_2w_2)(\gamma - w_2)$

Solve $\frac{dw2}{dt}|_{w_2 = 0+\alpha}$, where $\alpha$ is a positive pertubation.

$\frac{dw2}{dt}|_{w_2 = 0+\alpha} = x_2^2\alpha (\gamma-\alpha)$

For $\alpha < \gamma$, $\frac{dw2}{dt}|_{w_2 = 0+\alpha}$ is positive.

Therefore, given a pertubation $\alpha$ less than $\gamma$, the derivative $\frac{dw2}{dt}$ is postive and does not return to 0, so the fixed point $(w_1,w_2) = (0,0)$ is unstable.

Looking at the fixed point $(w_1,w_2) = (\gamma,\gamma)$:

1. 

Let $\frac{dw_1}{dt} = \gamma, \frac{dw2}{dt} = x_2(x_1w_1+x_2w_2)(\gamma - w_2)$

Solve $\frac{dw2}{dt}|_{w_2 = \gamma+\alpha}$, where $\alpha$ is a positive pertubation.

$\frac{dw2}{dt}|_{w_2 = 0+\alpha} = x_2(x_1\gamma + x_2\gamma+x_2\alpha)(-\alpha)$

This is negative since the first two terms are always positive and the last term is negative. This does not rule out a stable point, so we must keep checking.

2. 

Now, let $\alpha$ be a negative pertubation. Now, $x_2(x_1\gamma + x_2\gamma+x_2\alpha)(-\alpha)$ is positive. Once again, this does not rule out a stable point, so we must keep checking.

3. 

Let $\frac{dw_2}{dt} = \gamma, \frac{dw1}{dt} = x_1(x_1w_1+x_2w_2)(\gamma - w_1)$

Solve $\frac{dw1}{dt}|_{w_1 = \gamma+\alpha}$, where $\alpha$ is a positive pertubation.

$\frac{dw1}{dt}|_{w_1 = 0+\alpha} = x_1(x_2\gamma + x_1\alpha)(-\alpha)$

This is negative since the first two terms are always positive and the last term is negative. This does not rule out a stable point, so we must keep checking.

4. 

Now, let $\alpha$ be a negative pertubation. Now, $x_1(x_2\gamma + x_1\alpha)(-\alpha)$ is positive. Once again, this does not rule out a stable point, and we have checked all directions, so this point is stable.

To conclude, the fixed point $(w_1,w_2) = (0,0)$ is unstable, while the fixed point $(w_1,w_2) = (\gamma,\gamma)$ is stable.

This means that at $(w_1,w_2) = (0,0)$, there are no interactions, which is an equilibrium of the system. However, if a slight interaction forms (ie $w_1$ or $w_2$ become positive), this interaction will grow and not return to the fixed point of (0,0). However, at $(w_1,w_2) = (\gamma,\gamma)$ the interactinos between $x_1$ and $y$ and $x_2$ and $y$ are both equal to the bounded perameter $\gamma$. Small pertubations in any direction will not affect the system, as it will always come back to this stable fixed point as time t approaches $\infty$.

6. Write a Python script to simulate a Lotka-Volterra ODE model describing the interactions between a population of lynxes (L) and a population of hares (H). Simulate the dynamics of lynxes and hares for a time range of 50 days and plot the results as trajectories for lynxes and hares.

$\frac{dL}{dt} = cLH - d_LL$

$\frac{dH}{dt} = g_HH - cLH$

where $c$ represents rate of consumption, $d_L$ represents lynx death rate, and $g_H$ represents hare growth rate

a. Simulate the population dynamics for the following parameters: hare growth rate: 1.0 per day; rate of hare consumption by lynx: 0.5 per day; lynx death rate: 2 per day. Describe the results.

Plug in $c = 0.5$, $g_H = 1$, and $d_L = 2$

$\frac{dL}{dt} = 0.5LH - 2L$

$\frac{dH}{dt} = 1H - 0.5LH$

// rewritten for Euler's

$L(t) = L(t-1) + dt \times (0.5LH - 2L)$

$H(t) = H(t-1) + dt \times (1H - 0.5LH)$

```python
dt = 0.1
c = 0.5
g_H = 1
d_L = 2

hare_pop = []
lynx_pop = []

hare_pop.append(10)
lynx_pop.append(2)

for time_step in range (1, 50):
    hare_new = hare_pop[time_step-1] + dt*(g_H*hare_pop[time_step-1] - c*lynx_pop[time_step-1]*hare_pop[time_step-1])
    lynx_new = lynx_pop[time_step-1] + dt*(c*lynx_pop[time_step-1]*hare_pop[time_step-1] - d_L*lynx_pop[time_step-1])
    hare_pop.append(hare_new)
    lynx_pop.append(lynx_new)
    
plt.plot(range(0,50), lynx_pop)
plt.plot(range(0,50), hare_pop)
legend_text = ["Lynxes", "Hares"]
plt.legend(legend_text)
```

Plotting the hare and lynx populations with the starting populations H(0) = 10 and L(0) = 2, we can see a generally inverse relationship between hare and lynx populations. As the lynx population grows (feeding on hares), the hare population declines, and visa versa. We see that in this model, the lynx population runs out of food and does not recover, while the hare population runs rampant with no predator.
